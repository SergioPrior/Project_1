```python
# nos quedamos con la información que necesitamos de la tabla, en este caso la población por comunidad autonoma.
```

```python
poblacion = {"poblacion":[8600224, 1354144, 2224343, 2383702, 2089074, 7970295, 674796, 6917111, 5268012,
                         2699938, 323377, 2224072, 1560347 ]   
}
```

```python
columna = pd.DataFrame(poblacion) # creamos un dataframe para poder hacerle un join.
```

```python
df = df.join(columna)
```

```python
df
```

```python
victimas_por_cada_M_hab = df["n_victimas_mortales"] / df["poblacion"] * 1000000 

#sacamos victimas por cada millon de habitantes.
 
```

```python
data = pd.DataFrame(victimas_por_cada_M_hab)
```

```python
data.rename(columns={0:"victimas_por_cada_M_hab"}, inplace=True)
```

```python
df = df.join(data)
```

```python
df.n_victimas_mortales=df.n_victimas_mortales.astype(int)
```

```python
df.victimas_por_cada_M_hab=df.victimas_por_cada_M_hab.round(2)
```

```python
df.sort_values(by="victimas_por_cada_M_hab", ascending=True, inplace= True)
```

```python
plt.barh(df["comunidad_autonoma"], df["victimas_por_cada_M_hab"], color= "green", alpha= 0.4)


plt.xticks(rotation=80)
plt.xlabel("Nº de victimas por cada M de habitantes")
plt.title("Nº de casos en 2023: 57")

plt.show()
```
