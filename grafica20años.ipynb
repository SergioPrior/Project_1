```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("vic_cAut.csv")
```

```python
df=df.sort_values(by="n_victimas_mortales",ascending=False)
```

```python
plt.barh(df["comunidad_autonoma"], df["n_victimas_mortales"],color="blue")
plt.xticks(rotation=70)
plt.xlabel("Nº Muertes 2003-2023.")
plt.ylabel("Comunidad Autonoma.")
plt.title("Muertes por Violencia de genero por C.Autonoma.")

plt.show()
```

```python
df.drop("Unnamed: 0", axis = 1, inplace= True)
```

```python
df = df.reset_index()
df.drop(["index"], axis=1, inplace=True)
```
