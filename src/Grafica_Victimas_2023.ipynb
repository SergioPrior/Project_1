```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("vic_año-hab.csv")

df.n_victimas_mortales=df.n_victimas_mortales.astype(int)
df
```

```python
df=df.sort_values(by="n_victimas_mortales",ascending=False)
```

```python
plt.barh(df["comunidad_autonoma"], df["n_victimas_mortales"],color="blue", alpha=0.3)

for index, value in enumerate(df["n_victimas_mortales"]):
    plt.text(value, index, f'{value}', fontsize=8)

plt.xticks(rotation=70)
plt.xlabel("Nº Muertes 2023")
plt.title("Muertes por violencia de genero por cada C. Autonoma")

plt.show()
```

```python
df.drop("Unnamed: 0", axis = 1, inplace= True)
```
