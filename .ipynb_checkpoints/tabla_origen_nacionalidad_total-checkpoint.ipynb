```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("tabla_limpia.csv")
```

```python
tabla=df.groupby(["comunidad_autonoma","pais_nac_agresor"]).pais_nac_agresor.count().unstack().plot(kind="barh", width=0.8,figsize=(10, 6))
plt.xlabel("Nº muertes")
plt.ylabel("")
plt.title("Origen nacionalidad total muertes 2000-2023")
plt.show()
```
