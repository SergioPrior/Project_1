```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("tabla_limpia.csv")
df= df.sort_values(by="n_victimas_mortales", ascending=False)
```

```python
tabla=df.groupby(["comunidad_autonoma","pais_nac_agresor"]).pais_nac_agresor.count().unstack()
plt.xlabel("Nº muertes")
plt.ylabel("")
plt.title("Origen nacionalidad total muertes 2000-2023")
plt.show()
```

```python
type(tabla)
```

```python
tabla=tabla.fillna(0)
```

```python
tabla=tabla.sort_values(by="Otro país", ascending=False)
```

```python
tabla.plot(kind='barh', width=0.7)
```
