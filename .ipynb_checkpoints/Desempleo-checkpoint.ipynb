```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://datosmacro.expansion.com/paro-epa/espana-comunidades-autonomas"

desempleo = pd.read_html(url)[0]
```

```python
desempleo.drop(0, inplace=True)
```

```python
desempleo= desempleo.reset_index()
```

```python
desempleo.drop(columns=["Tasa de desempleo.1","Var.","Var. Año","Mes"], inplace = True)
```

```python
desempleo.CCAA=desempleo.CCAA.str.replace("[+]", "")
```

```python
desempleo.drop(columns="index", inplace=True)
```

```python
desempleo.rename(columns={"Tasa de desempleo":"tasa de desempleo %"}, inplace=True)
```

```python
desempleo.drop(19, inplace=True)
```

```python
desempleo["tasa de desempleo %"]=desempleo["tasa de desempleo %"].str.replace("%","").str.replace(",",".")
```

```python
desempleo["tasa de desempleo %"]=desempleo["tasa de desempleo %"].astype(float)
```

```python
desempleo.sort_values(by="tasa de desempleo %", ascending=True, inplace =True)
```

```python
desempleo.to_csv("desempleo.csv", index = False)
desempleo
```

```python
plt.barh(desempleo["CCAA"], desempleo["tasa de desempleo %"], color= "black", alpha= 0.4)

for index, value in enumerate(desempleo["tasa de desempleo %"]):
    plt.text(value, index, f'{value:.2f}', fontsize=8)

plt.xticks(rotation=80)
plt.xlabel("Porcentaje de desemeplo diciembre 2023")
plt.title("DESEMPLEO ESPAÑA ACTUALIZACIÓN DICIEMBRE 2023")

plt.show()
```
